```python
import nest_asyncio
import os
import asyncio
from telegram import Update
from telegram.ext import ApplicationBuilder, CommandHandler, MessageHandler, filters

# Patch the running event loop in Jupyter
print("Applying nest_asyncio patch...")
nest_asyncio.apply()
print("Patch applied.\n")

# Handler for the /start command with print statements
async def start(update: Update, context):
    print("Received /start command.")
    await update.message.reply_text("Niaje Andre my Huzy! Kama your Bae,Send me an image, and I'll save it.")
    print("Replied to /start command.\n")

# Handler for incoming photos with print statements
async def handle_photo(update: Update, context):
    print("Received a photo message.")
    
    # Get the highest resolution photo
    photo = update.message.photo[-1]
    file_id = photo.file_id
    print(f"Extracted file_id: {file_id}")
    
    # Get file information from Telegram
    new_file = await context.bot.get_file(file_id)
    print("Retrieved file information from Telegram.")
    
    # Ensure the target directory exists
    if not os.path.exists('downloaded_images'):
        print("Directory 'downloaded_images' does not exist. Creating directory...")
        os.makedirs('downloaded_images')
        print("Directory created.")
    else:
        print("Directory 'downloaded_images' already exists.")
    
    # Define the path for saving the image
    file_path = os.path.join('downloaded_images', f'{file_id}.jpg')
    print(f"Saving image to: {file_path}")
    
    # Download and save the image
    await new_file.download_to_drive(file_path)
    print("Image downloaded and saved.")
    
    await update.message.reply_text(f"Image has been saved to {file_path}")
    print("Replied to user after saving image.\n")

# Asynchronous main function to set up and run the bot
async def async_main():
    print("Building the application...")
    # Replace "YOUR_BOT_TOKEN" with your actual bot token
    application = ApplicationBuilder().token(BOT_TOKEN).build()
    print("Application built.")

    # Register handlers with print statements
    print("Registering /start command handler...")
    application.add_handler(CommandHandler("start", start))
    print("Registering photo message handler...")
    application.add_handler(MessageHandler(filters.PHOTO, handle_photo))
    print("Handlers registered.\n")
    
    # Initialize and start the bot
    print("Initializing application...")
    await application.initialize()
    print("Application initialized.")
    
    print("Starting application...")
    await application.start()
    print("Application started.")
    
    print("Starting polling for updates...")
    await application.updater.start_polling()
    print("Polling started.\n")
    
    print("Bot is now running and idle. Press CTRL+C to stop.")
    # Keep the bot running indefinitely
    await asyncio.Event().wait()

# Run the bot in a Jupyter Notebook cell
print("Starting async_main()...")
await async_main()
```

```python
import os
from telegram import Update
from telegram.ext import ApplicationBuilder, CommandHandler, MessageHandler, filters

# Handler for the /start command
async def start(update: Update, context):
    print("Received /start command.")
    await update.message.reply_text("Hello! Send me an image, and I'll save it.")
    print("Replied to /start command.\n")

# Handler for incoming photos
async def handle_photo(update: Update, context):
    print("Received a photo message.")
    
    # Get the highest resolution photo
    photo = update.message.photo[-1]
    file_id = photo.file_id
    print(f"Extracted file_id: {file_id}")
    
    # Retrieve file info from Telegram
    new_file = await context.bot.get_file(file_id)
    print("Retrieved file information from Telegram.")
    
    # Ensure the target directory exists
    if not os.path.exists('downloaded_images'):
        print("Directory 'downloaded_images' does not exist. Creating directory...")
        os.makedirs('downloaded_images')
        print("Directory created.")
    else:
        print("Directory 'downloaded_images' already exists.")
    
    # Define the path for saving the image
    file_path = os.path.join('downloaded_images', f'{file_id}.jpg')
    print(f"Saving image to: {file_path}")
    
    # Download and save the image
    await new_file.download_to_drive(file_path)
    print("Image downloaded and saved.")
    
    await update.message.reply_text(f"Image has been saved to {file_path}")
    print("Replied to user after saving image.\n")

def main():
    print("Building the application...")
    # Replace "YOUR_BOT_TOKEN" with your actual bot token
    application = ApplicationBuilder().token("YOUR_BOT_TOKEN").build()
    print("Application built.")
    
    # Register handlers
    print("Registering handlers...")
    application.add_handler(CommandHandler("start", start))
    application.add_handler(MessageHandler(filters.PHOTO, handle_photo))
    print("Handlers registered.\n")
    
    print("Starting polling for updates. Bot is running...")
    application.run_polling()
    print("Polling stopped.")

if __name__ == '__main__':
    main()
```
